```python
# Import modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import string
import re
import warnings
warnings.filterwarnings("ignore")
import os

# Get the current working directory
current_dir = os.getcwd()
parent_dir = os.path.abspath(os.path.join(current_dir, os.pardir))
```

```python
# From signature analysis, import the weighted probabilities of indels and stop-gained mutations in each region of APC

signature_corrected=pd.read_csv(parent_dir+'/Signature Analysis/Results/All_signatures_weights.csv')
```

```python

# Find weight per region for stop_gained mutations
L=np.asarray(signature_corrected['SBS_healthy']) # for MSS (healthy crypts data)
L_POLE=np.asarray(signature_corrected['SBS_POLE']) # for POLE
L_MSI=np.asarray(signature_corrected['SBS_MSI']) # for MSI

# Find weight per region for stop_gained mutations
I=np.asarray(signature_corrected['IDS_healthy']) # for MSS (healthy crypts data)
I_MSI=np.asarray(signature_corrected['IDS_MSI']) # for MSI

fs=3/7 # Probabilty of frameshifts, MSS
fs_MSI=9/14 # Probabilty of frameshifts, MSI

# % of tumors with genotyped [0,-], [0,0] and [0,x2]
dels=44
dms=23
cnlohs=33
```

```python

         
# Function to determine the total number of 20AARs

def aars(pr):
    if len(pr)==2:    # and 
        if pr[1]==0.5:
            aar=pr[0]*2
        else:
            aar=sum(pr)
        
    else:
        aar=sum(pr)
    return aar

# Function to calculate the mutation probabilities

def NULL(L,I,fs,dels,cnlohs):
    L0=L[0]
    L1=L[1]
    L2=L[2]
    L3=L[3]

    I0=I[0]
    I1=I[1]
    I2=I[2]
    I3=I[3]

    m0=1
    
    # relative to m0
    
    m1=(L1*fs+(1-fs)*I1) /(L0*fs+(1-fs)*I0)
    m2=(L2*fs+(1-fs)*I2 )/(L0*fs+(1-fs)*I0)
    m3=(L3*fs+(1-fs)*I3 )/(L0*fs+(1-fs)*I0)

    md=(dels/dms)*0.5
    mn=(cnlohs/dms)
    

    total=(mn+md+1+m1+m2+m3)
    t=total

    m1=m1/t
    m2=m2/t
    m0=1/t
    m3=m3/t
    mn=mn/t
    md=md/t


    
    m00=m0**2
    m11=m1**2
    m22=m2**2
    m33=m3**2
    
    m01=2*m1*m0
    m02=2*m2*m0
    m03=2*m0*m3
    
    m12=2*m1*m2
    m23=2*m2*m3
    m13=2*m1*m3
    
    md0=2*m0*md
    md1=2*m1*md
    md2=2*m2*md
    md3=2*m3*md
    
    mn0=m0*mn
    mn1=m1*mn
    mn2=m2*mn
    mn3=m3*mn
    

    
    Types_double=[[0,0],[1,1],[2,2],[3,3],[0,1],[1,2],[1,3],[0,2],[2,3],[0,3],[0],[1],[2],[3],[0,0.5],[1,0.5],[2,0.5],[3,0.5]]
    aar=[]
    for t in Types_double:
        aar.append(aars(t))
        
    Types_double=['[0, 0]','[1, 1]','[2, 2]','[3, 3]','[0, 1]','[1, 2]','[1, 3]','[0, 2]','[2, 3]','[0, 3]','[0]','[1]','[2]','[3]','[0, x2]','[1, x2]','[2, x2]','[3, x2]']
    probs_double=np.asarray([m00, m11, m22, m33, m01, m12, m13, m02, m23, m03,md0,md1,md2,md3, mn0, mn1, mn2,mn3])

    
    
    

    return pd.DataFrame(data={'Genotype':Types_double,'20AARs':aar ,'Probabilities':probs_double/sum(probs_double)})#,pd.DataFrame(data={'Genotype':types,'Probabilities':probs_s/sum(probs_s)})
```

```python
null_mss=NULL(L,I,fs,dels,cnlohs)
null_pole=NULL(L_POLE,I,0,0,0)
null_MSI=NULL(L_MSI,I_MSI,fs_MSI,0,0)

```

```python
# Save results

null_pole.to_csv('Results/MutationProbabilities_POLE.csv')
null_mss.to_csv('Results/MutationProbabilities_MSS.csv')
null_MSI.to_csv('Results/MutationProbabilities_MSI.csv')
```
